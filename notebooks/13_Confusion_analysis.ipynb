```python
import pandas as pd

cm = pd.read_csv("confusion_matrix.csv", index_col=0)
```

*Extract Misclassifications*

only want off-diagonal values 

```python
confusions = []

for true_cwe in cm.index:
    total_samples = cm.loc[true_cwe].sum()

    for pred_cwe in cm.columns:

        if true_cwe != pred_cwe:

            count = cm.loc[true_cwe, pred_cwe]

            if count > 0:

                confusion_rate = count / total_samples

                confusions.append({
                    "True_CWE": true_cwe,
                    "Predicted_CWE": pred_cwe,
                    "Count": count,
                    "Confusion_Rate": confusion_rate
                })
confusions_df = pd.DataFrame(confusions)

confusions_df = confusions_df.sort_values(
    "Count",
    ascending=False
)

print(confusions_df.head(20))
confusions_df.to_csv(
    "cwe_confusion_analysis.csv",
    index=False
)
```

Only keep the important ones

```python
critical_confusions = confusions_df[
    confusions_df["Confusion_Rate"] > 0.05
]

print(critical_confusions)
```
